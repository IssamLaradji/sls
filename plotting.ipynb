```python
import pprint
import utils as ut
import pandas as pd
import pylab as plt

%matplotlib inline

fname = "CIFAR10-ResNet34-softmax"
plt.title(fname)

for exp_id in ["2b7f4319fe31daa661f7e831b9550855",
               "ded7fe7a05f009721b141ebb0bce2ca4"]:
    exp_dict = ut.load_yaml("results/%s/exp_dict.yaml" % exp_id)
    history = ut.load_pkl("results/%s/history.pkl" % exp_id)

    pprint.pprint(exp_dict)

    x_list = pd.DataFrame(history["scoreList"])["iters"].cumsum()
    y_list = pd.DataFrame(history["scoreList"])["test_acc"]


    plt.plot(x_list[10:], y_list[10:], label=exp_dict["opt"])
    
plt.legend()
plt.savefig("results/%s.png" % fname, format = "png", 
            dpi = 1200, bbox_inches="tight")
plt.show()
```
